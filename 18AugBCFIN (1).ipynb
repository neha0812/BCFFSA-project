```python
import numpy as np 
import pandas as pd
sample=pd.read_csv(r"C:\Users\neha4\Desktop\PIMS_2020_Real_Estate_data-master\PIMS_2020_Real_Estate_data-master\sample_clean.csv")
print("successfully read the file")
sample2=sample
sum_column=(sample2["saf1"]+sample2["saf2"]+sample2["saf3"]+sample2["saf4"]+sample2["saf5"]+sample2["saf6"]+sample2["saf7"]+sample2["saf8"])/8
sample2["av_saf2"]=sum_column
sample2=sample2.sort_values(by='av_saf2',ascending=False)
sample2.dropna(subset = ["saf1"], inplace=True)
sample2.dropna(subset = ["saf2"], inplace=True)
sample2.dropna(subset = ["saf3"], inplace=True)
sample2.dropna(subset = ["saf4"], inplace=True)
sample2.dropna(subset = ["saf5"], inplace=True)
sample2.dropna(subset = ["saf6"], inplace=True)
sample2.dropna(subset = ["saf7"], inplace=True)
sample2.dropna(subset = ["saf8"], inplace=True)
sample2.dropna(subset = ["YEAR_OF_CONSTRUCTION"], inplace=True)
sample2.dropna(subset = ["pct_change"], inplace=True)
sample2.dropna(subset = ["mr5y"], inplace=True)
sample2.dropna(subset = ["ASSESSED_VALUE"], inplace=True)
```

```python
A= sample2[['YEAR_OF_CONSTRUCTION','mr5y','saf1','saf2','saf3','saf4','saf5','saf6','saf7','saf8']]
B = sample2["ASSESSED_VALUE"]
new_header = A.iloc[0] #grab the first row for the header
A = A[1:] #take the data less the header row
B.columns = new_header #set the header row as the df header
new_header = B.iloc[0] #grab the first row for the header
B = B[1:] #take the data less the header row
B.columns = new_header #set the header row as the df header
X = A
y = B
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn import datasets, linear_model, metrics 
  

# defining feature matrix(X) and response vector(y) 

X = A
y = B
# splitting X and y into training and testing sets 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, 
                                                    random_state=1) 
  
# create linear regression object 
reg = linear_model.LinearRegression() 
  
# train the model using the training sets 
reg.fit(X_train, y_train) 
  
# regression coefficients 
print('Coefficients: \n', reg.coef_) 
  
# variance score: 1 means perfect prediction 
print('Variance score: {}'.format(reg.score(X_test, y_test))) 
  
# plot for residual error 
  
## setting plot style 
plt.style.use('fivethirtyeight') 
plt.figure(figsize=(40,20))
  
## plotting residual errors in training data 
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, 
            color = "green", s = 10, label = 'Train data') 
  
## plotting residual errors in test data 
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, 
            color = "blue", s = 10, label = 'Test data') 
  
## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2) 
#fig, ax = plt.subplots(figsize=(20,4))
  
## plotting legend 
plt.legend(loc = 'upper right') 
#plt.figure(figsize=(20,20))  
## plot title 
plt.title("Residual Error") 

plt.savefig('plot.png', dpi=300, bbox_inches='tight')
  
## function to show plot 
plt.show() 
```
